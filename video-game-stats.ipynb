```python
df = pd.read_csv('/kaggle/input/videogamesales/vgsales.csv') # Reads in the dataset to the pandas dataframe
df_reset = df.set_index("Rank") # remove the index column and set the index to be the rank
df_reset
```

### 1. Which company is the most common video game publisher?

```python
most_common_publisher = df_reset.Publisher.mode()[0]
print("The most common publisher of video games is", most_common_publisher)
```

### 2. What is the most common platform?

```python
most_common_platform = df_reset.Platform.mode()[0]
print("The most common platform is the", most_common_platform)
```

### 3. What is the most common genre?

```python
most_common_genre = df_reset.Genre.mode()[0]
print("The most common genre of video games is", most_common_genre)
```

### 4. What are the top 20 highest grossing games?

```python
sorted_global_sales = df_reset.sort_values(by="Global_Sales")
top20_global_sales = df_reset.head(20)
top20_global_sales
```

### 5. For North American video game sales, what is the median?

```python
na = df_reset["NA_Sales"]
na_median_sales = na.median()
print("The median North American average sales is", na_median_sales)
```

* Provide a secondary output showing 10 games surrounding the median sales output
* Assume that games with same median value are sorted in descending order

```python
na_games_around_median = df_reset[ df_reset['NA_Sales'] == na_median_sales]
descend_na_games_around_median = na_games_around_median.sort_values(by="Name", ascending=True)
descend_na_games_around_median.head(10)
```

### 6. For the top-selling game of all time, how many standarad deviations above/below the mean are its sales for North America?

```python
std_na_sales = df_reset["NA_Sales"].std()
rounded_std_na_sales = round((std_na_sales), 4)
print("The top selling game of all time is " + str(rounded_std_na_sales) + " above sales for North America.")
```

### 7. The Nintendo Wii seems to have outdone itself with games. How does its average number of sales compare with all of the other platforms?

```python
wii_platform = df_reset[ df_reset['Platform'] == 'Wii']
wii_avg = wii_platform['Global_Sales'].mean()
all_other_platforms = df_reset[ df_reset['Platform'] != 'Wii']
all_other_platforms_avg = all_other_platforms['Global_Sales'].mean()
diff_in_wii_vs_others = round(((wii_avg / all_other_platforms_avg) * 100), 2)
print("The Wii outperforms all other platforms by", diff_in_wii_vs_others, "%")
```

### 8. come up with 3 more questions that can be answered by this data set.
* How many games did Wii release in 2006?

```python
wii_2006_year = wii_platform[ wii_platform['Year'] == 2006]
wii_2006_year_sales = wii_2006_year['Global_Sales'].sum()
print("Global sales for Wii in 2006 totalled $" + str(wii_2006_year_sales) + " million dollars.")
```

* What was the earliest year used in the data set?

```python
earliest_reported_year = df_reset.Year.min()
formatted_earliest_reported_year = str(int(earliest_reported_year))
print("This data set goes as far back as far as " + formatted_earliest_reported_year + '.')
```
